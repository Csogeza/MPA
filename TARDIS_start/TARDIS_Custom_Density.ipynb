```python
# Run simulation

sim = run_tardis('tardis_custom.yml')
```

```python
sim2 = run_tardis('tardis_example_powerlaw.yml')
```

```python
def plot_spectrum(sim):
    spectrum = sim.runner.spectrum
    spectrum_virtual = sim.runner.spectrum_virtual
    spectrum_integrated = sim.runner.spectrum_integrated

    plot(spectrum.wavelength, spectrum.luminosity_density_lambda, label='normal packets')
    plot(spectrum.wavelength, spectrum_virtual.luminosity_density_lambda, label='virtual packets')
    plot(spectrum.wavelength, spectrum_integrated.luminosity_density_lambda, label='formal integral')

    xlabel('Wavelength [$\AA$]')
    ylabel('Luminosity [erg/s/$\AA$]')
    legend()
    xlim(3000,9000)
    ylim(-0.5e39,6e39)
```

```python
def get_spectra(sim):
    spectrum = sim.runner.spectrum
    spectrum_virtual = sim.runner.spectrum_virtual
    spectrum_integrated = sim.runner.spectrum_integrated
    
    return spectrum, spectrum_virtual, spectrum_integrated
```

```python
f = figure(figsize=(16,6))
ax = f.add_subplot(1,2,1)
plot_spectrum(sim)
ax = f.add_subplot(1,2,2)
plot_spectrum(sim2)
```

```python
s1, sv1, si1 = get_spectra(sim)
s2, sv2, si2 = get_spectra(sim2)
```

```python
figure(figsize=(8,6))

plot(s1.wavelength, s1.luminosity_density_lambda - s2.luminosity_density_lambda)
plot(s1.wavelength, sv1.luminosity_density_lambda - sv2.luminosity_density_lambda)
plot(s1.wavelength, si1.luminosity_density_lambda - si2.luminosity_density_lambda)

xlim(3000,9000)
```
